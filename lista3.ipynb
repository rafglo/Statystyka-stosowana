```python
dyrektor = [41,27,35,33,25,47,38,53,43,35,36]
wizytator = [38,24,34,29,27,47,43,52,39,31,29]
```

```python
def sprawdzanie_powtorzen(wartosci):
    return [wartosci.count(wartosci[i]) for i in range(len(wartosci))]
```

```python
print(sprawdzanie_powtorzen(dyrektor))
print(sprawdzanie_powtorzen(wizytator))
```

```python

def nadawanie_rang(wartosci):
    wartosci_sorted = wartosci.copy()
    wartosci_sorted.sort(reverse=True)
    rangi = [0]*len(wartosci)
    for i in range(len(wartosci_sorted)):
        ile = wartosci.count(wartosci_sorted[i]) 
        if ile == 1:
            ind = wartosci.index(wartosci_sorted[i])
            rangi[ind] = i + 1
        else:
            ind_sum = ile
            inds = []
            for j in range(len(wartosci)):
                if wartosci_sorted[i] == wartosci[j]:
                    inds.append(j)

            for k in range(len(wartosci_sorted)):
                if wartosci_sorted[i] == wartosci_sorted[k]:
                    ind_sum += k
                    
            ranga = ind_sum/ile
            for ind in inds:
                rangi[ind] = ranga
    return rangi
```

```python
def spearman(wartosci1, wartosci2, czy_rangi = False):
    if czy_rangi:
        rangi1, rangi2 = wartosci1, wartosci2
    else:
        rangi1, rangi2 = nadawanie_rang(wartosci1), nadawanie_rang(wartosci2)
    n = len(wartosci1)
    sum_d = 0
    for i in range(len(rangi1)):
        sum_d += (rangi1[i] - rangi2[i])**2
    result = 1 - 6 * sum_d/(n*(n**2-1))
    return result
```

```python
print(spearman(dyrektor, wizytator))
```

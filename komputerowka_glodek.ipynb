```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
def simul(n, theta):
    u = np.random.rand(n)
    return u**(theta)
```

```python
sample = simul(1000, 8)
```

```python
def teo_cdf(x, theta):
    return x**(1/theta)
```

```python
xs = np.linspace(0,1,1000)
theta = 8
```

```python
plt.figure(figsize=(12,6))
sns.ecdfplot(sample, label="Dystrybuanta empiryczna")
plt.plot(xs, teo_cdf(xs, theta), label="Dystrybuanta teoretyczna")
plt.legend(loc="best")
plt.show()
```

```python
def theta_estim(sample):
    n = len(sample)
    return -np.sum(np.log(sample))/n
```

```python
def theta_estim_moments(sample):
    sam_mean = np.mean(sample)
    return 1/sam_mean - 1
```

```python
n = 1000
n_samples = 100
theta = 8
samples_ = [simul(n_samples, theta) for n in range(n_samples)]
thetas1 = [theta_estim(sam) for sam in samples_]
thetas2 = [theta_estim_moments(sam) for sam in samples_]
plt.figure(figsize=(15,9))
plt.boxplot([thetas1, thetas2], labels=["Metoda NW", "Metoda momentów"])
plt.title("Boxplot estymatorów theta")
plt.show()
```

```python
def u_cdf(theta, x, n):
    return x**(n/theta)
```

```python
def u_simul(n, theta):
    u = np.random.rand(n)
    return u**(theta/n)
```

```python
sample1 = u_simul(1000, 8)
```

```python
xs2 = np.linspace(0.1,1,1000)
```

```python

plt.figure(figsize=(15,9))
sns.ecdfplot(sample1, label="Dystrybuanta empiryczna")
plt.plot(xs2, u_cdf(8, xs2, 1000), label="Dystrybuanta teoretyczna")
plt.legend(loc="best")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def simul(n, theta):
    u = np.random.rand(n)
    return u**(theta)
```

```python
sample = simul(1000, 8)
```

```python
def teo_cdf(x, theta):
    return x**(1/theta)
```

```python
xs = np.linspace(0,1,1000)
theta = 8
```

```python
plt.figure(figsize=(12,5))
sns.ecdfplot(sample, label="Dystrybuanta empiryczna")
plt.plot(xs, teo_cdf(xs, theta), label="Dystrybuanta teoretyczna")
plt.legend(loc="best")
plt.show()
```

```python
def theta_estim(sample):
    n = len(sample)
    return -np.sum(np.log(sample))/n
```

```python
n = 1000
n_samples = 100
theta = 8
samples_ = [simul(n_samples, theta) for n in range(n_samples)]
thetas = [theta_estim(sam) for sam in samples_]
plt.figure(figsize=(15,9))
plt.boxplot(thetas)
plt.title("Boxplot estymatora theta")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def dystrybuanta_x(x, theta):
    return (1-theta**2/x**2)
```

```python
def dystrybuanta_u(x, theta, n):
    return (1-theta**2/x**2)**n
```

```python
def u_odworotna_dys(n, theta):
    u = np.random.rand(n)
    return np.sqrt(theta**2/(1-u**(1/n)))
```

```python
sample = u_odworotna_dys(1000, 5)
```

```python
xs = np.linspace(0.1,2000,1000)
```

```python

plt.figure(figsize=(15,9))
sns.ecdfplot(sample, label="Dystrybuanta empiryczna")
plt.plot(xs, dystrybuanta_u(xs, 5, 1000), label="Dystrybuanta teoretyczna")
plt.legend(loc="best")
plt.title("Porównanie dystrybuanty empirycznej i teoretycznej")
plt.xlim(0,2000)
plt.show()
```

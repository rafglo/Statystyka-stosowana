```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python

Y = pd.read_csv("rym_top_5000_all_time1.csv")['Average Rating'].tolist()
dates = pd.read_csv("rym_top_5000_all_time1.csv")['Release Date'].tolist()
years = []
for date in dates:
    years.append(int(date[-4:]))

X = np.array(years).reshape(-1, 1)
```

```python
model = LinearRegression() #tworzymy model regresji i pniej trzeba go dopasowa dlatego uywam fit()
model.fit(X, Y) #to jest najlepsze dopasowanie prostej do danych 
Y_pred = model.predict(X)


plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='plum', label='Dane') 
plt.plot(X, Y_pred, color='purple', linewidth=2, label='Regresja liniowa')  
plt.title('Release Date vs Average Rating')
plt.xlabel('Release Date', fontsize=11, color= 'darkmagenta')
plt.ylabel('Average Rating',fontsize=11,  color= 'darkmagenta')
plt.gca().set_facecolor('snow')
plt.legend()
plt.show()
```

```python
r_sq = model.score(X, Y)
print(f"coefficient of determination: {r_sq}")
#wspczynnik determinacji mwi jak dobrze model regresji dopasowuje si do danych. Im  bliej 1 tym model lepiej sie dopasowduje do danych
```

```python
print(f"intercept: {model.intercept_}") #b
print(f"slope: {model.coef_}") #a
```

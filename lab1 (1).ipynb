```python
import matplotlib.pyplot as plt
import math
import numpy as np

def dystrybuanta(x, wsp):
    return 1 - math.e**(-wsp * x)

xs = np.linspace(0,6,1000)
ys1 = [dystrybuanta(i, 0.5) for i in xs]
ys2 = [dystrybuanta(i, 1) for i in xs]
ys3 = [dystrybuanta(i, 5) for i in xs]
plt.plot(xs, ys1, label="lamba=0.5")
plt.plot(xs, ys2, label='lamba=1')
plt.plot(xs, ys3, label='lambda=5')
plt.xlabel("x")
plt.ylabel("F(x)")
plt.legend(loc="best")
plt.title("Dystybuanta rozkładu wykładniczego dla różnych parametrów")
plt.show()

```

```python
import seaborn as sns
probka = np.random.exponential(2, 1000)
xs2 = np.linspace(0,14)
ys4 = [dystrybuanta(i, 2) for i in xs2]
sns.ecdfplot(data=probka, label="empriyczna")
plt.plot(ys4, label='teoretyczna')
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.axis
plt.show()
```

```python
import numpy as np
import scipy.stats
```

```python
n1 = 20
n2 = 50
n3 = 100
n = 1000
mi = 2.1
alpha = 0.02
v = 0.2
std = np.sqrt(v)
```

```python
def confidence_interval_mean_known_variance(sample, variance, alpha):
    std = np.sqrt(variance)
    q = scipy.stats.norm.ppf(1-alpha/2)
    x_mean = np.mean(sample)
    n = len(sample)
    return x_mean - q * std/np.sqrt(n), x_mean + q*std/np.sqrt(n)
```

```python
def confidence_interval_mean_unknown_variance(sample, alpha):
    n = len(sample)
    x_mean = np.mean(sample)
    x_std = np.std(sample)
    q = scipy.stats.t.ppf(1-alpha/2, n-1)
    return x_mean - q * x_std/np.sqrt(n), x_mean + q * x_std/np.sqrt(n)
```

```python
def confidence_interval_variance(sample, alpha):
    n = len(sample)
    variance = np.var(sample)
    q1, q2 = scipy.stats.chi.ppf([1-alpha/2, alpha/2], n-1)
    return (n-1) * variance/q1, (n-1) * variance/q2
```

```python
mi_in_conf_int_n1 = 0
mi_in_conf_int_n2 = 0
mi_in_conf_int_n3 = 0
mi_in_conf_int2_n1 = 0
mi_in_conf_int2_n2 = 0
mi_in_conf_int2_n3 = 0
for i in range(n):
    sample_n1 = np.random.normal(mi, std, n1)
    sample_n2 = np.random.normal(mi, std, n2)
    sample_n3 = np.random.normal(mi, std, n3)
    conf_int_n1 = confidence_interval_mean_known_variance(sample_n1, v,alpha)
    conf_int_n2 = confidence_interval_mean_known_variance(sample_n2, v,alpha)
    conf_int_n3 = confidence_interval_mean_known_variance(sample_n3, v,alpha)
    conf_int2_n1 = confidence_interval_mean_unknown_variance(sample_n1,alpha)
    conf_int2_n2 = confidence_interval_mean_unknown_variance(sample_n2,alpha)
    conf_int2_n3 = confidence_interval_mean_unknown_variance(sample_n3,alpha)
    if conf_int_n1[0] <= mi <= conf_int_n1[1]:
        mi_in_conf_int_n1 += 1 
    if conf_int_n2[0] <= mi <= conf_int_n2[1]:
        mi_in_conf_int_n2 += 1 
    if conf_int_n3[0] <= mi <= conf_int_n3[1]:
        mi_in_conf_int_n3 += 1 
    if conf_int2_n1[0] <= mi <= conf_int2_n1[1]:
        mi_in_conf_int2_n1 += 1 
    if conf_int2_n2[0] <= mi <= conf_int2_n2[1]:
        mi_in_conf_int2_n2 += 1 
    if conf_int2_n3[0] <= mi <= conf_int2_n3[1]:
        mi_in_conf_int2_n3 += 1 
print(f" Ile dla \nwariancji znanej \nn1 = 20: {mi_in_conf_int_n1},\nn2 = 50: {mi_in_conf_int_n2},\nn3 = 100: {mi_in_conf_int_n3} \n wariancji nieznanej \nn1 = 20: {mi_in_conf_int2_n1},\nn2 = 50: {mi_in_conf_int2_n2},\nn3 = 100: {mi_in_conf_int2_n3}")
```

```python
sigma_in_conf_int_n1 = 0
sigma_in_conf_int_n2 = 0
sigma_in_conf_int_n3 = 0
for i in range(n):
    sample_n1 = np.random.normal(mi, std, n1)
    sample_n2 = np.random.normal(mi, std, n2)
    sample_n3 = np.random.normal(mi, std, n3)
    conf_int_n1 = confidence_interval_variance(sample_n1,alpha)
    conf_int_n2 = confidence_interval_variance(sample_n2,alpha)
    conf_int_n3 = confidence_interval_variance(sample_n3,alpha)
    if conf_int_n1[0] <= v <= conf_int_n1[1]:
        sigma_in_conf_int_n1 += 1 
    if conf_int_n2[0] <= v <= conf_int_n2[1]:
        sigma_in_conf_int_n2 += 1 
    if conf_int_n3[0] <= v <= conf_int_n3[1]:
        sigma_in_conf_int_n3 += 1 
print(f" Ile dla \nwariancji znanej \nn1 = 20: {mi_in_conf_int_n1},\nn2 = 50: {mi_in_conf_int_n2},\nn3 = 100: {mi_in_conf_int_n3} \n wariancji nieznanej \nn1 = 20: {mi_in_conf_int2_n1},\nn2 = 50: {mi_in_conf_int2_n2},\nn3 = 100: {mi_in_conf_int2_n3}")
```
